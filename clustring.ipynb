```python
import random 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans 
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('/content/1632560262896716.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
X = df[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
from sklearn.cluster import KMeans
```

## **KMeans**

```python
 kmeans = KMeans(n_clusters= 5 , max_iter=500 , n_init=10)
 kmeans.fit_predict(X)
```

```python
kmeans.labels_
```

```python
plt.scatter(X['Annual Income (k$)'],X['Spending Score (1-100)'])
```

```python
plt.scatter(X['Annual Income (k$)'],X['Spending Score (1-100)'],c=kmeans.labels_ , cmap='rainbow')
```

```python
sse = []
for k in range(1,10):
  kmeans = KMeans(n_clusters=k)
  kmeans.fit(X)
  sse.append(kmeans.inertia_)
```

```python

plt.plot(range(1,10) , sse)
plt.xticks(range(1,10))
plt.xlabel("number of clusters")
plt.ylabel("SSE")
plt.show()
```

## **hierarchy**

```python
import scipy.cluster.hierarchy as sch
```

```python
data = X.values
data[:5]
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler()
data = min_max_scaler.fit_transform(data)
```

```python
clus1 = sch.linkage(data)
```

```python
plt.figure(figsize=(20,8))
dend = sch.dendrogram(clus1)
```

```python
from sklearn.cluster import AgglomerativeClustering
clus2 = AgglomerativeClustering(n_clusters=5 , linkage="single")
```

```python
clus2.fit(data)
```

```python
plt.figure(figsize=(10,7))
plt.scatter(data[:,0],data[:,1],c=clus2.labels_)
```

```python
from sklearn.cluster import AgglomerativeClustering
clus2 = AgglomerativeClustering(n_clusters=5 , linkage="complete")
```

```python
clus2.fit(data)
```

```python
plt.figure(figsize=(10,7))
plt.scatter(data[:,0],data[:,1],c=clus2.labels_)
```
